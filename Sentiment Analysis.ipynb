# Implementation

```python
x=df.drop(['30. Are you happy with COVID-19, RUSSIA-UKRAINE WAR, AND ISRAEL-HAMAS CONFLICTS respecting future or current Bangladeshi economic concerns?'],axis=1)
```

```python
x
```

```python
y=df['30. Are you happy with COVID-19, RUSSIA-UKRAINE WAR, AND ISRAEL-HAMAS CONFLICTS respecting future or current Bangladeshi economic concerns?']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
```

```python
xtrain,xtest,ytrain,ytest= train_test_split(x,y,test_size=.20,random_state=1)
```

```python
xtrain
```

# SVC

```python
model= SVC(gamma='auto')
```

```python
model.fit(xtrain,ytrain)
```

```python
model.score(xtest,ytest)
```

# Confusion Matrix for SVC

```python
# Creating SVC Model
from sklearn.svm import SVC
classifier = SVC()
classifier.fit(xtrain, ytrain)
```

```python
# Creating a confusion matrix
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
y_pred = classifier.predict(xtest)
cm = confusion_matrix(ytest, y_pred)
cm
```

```python
# Plotting the confusion matrix
plt.figure(figsize=(10,7))
p = sns.heatmap(cm, annot=True, cmap="Blues", fmt='g')
plt.title('Confusion matrix for Support Vector Classifier Model - Test Set')
plt.xlabel('Predicted Values')
plt.ylabel('Actual Values')
plt.show()
```

```python
# Accuracy Score
score = round(accuracy_score(ytest, y_pred),4)*100
print("Accuracy on test set: {}%".format(score))
```

```python
# Classification Report
print(classification_report(ytest, y_pred))
```

```python
# Creating a confusion matrix for training set
y_train_pred = classifier.predict(xtrain)
cm = confusion_matrix(ytrain, y_train_pred)
cm
```

```python
# Plotting the confusion matrix
plt.figure(figsize=(10,7))
p = sns.heatmap(cm, annot=True, cmap="Blues", fmt='g')
plt.title('Confusion matrix for SVC Model - Train Set')
plt.xlabel('Predicted Values')
plt.ylabel('Actual Values')
plt.show()
```

```python
# Accuracy Score
score = round(accuracy_score(ytrain, y_train_pred),4)*100
print("Accuracy on trainning set: {}%".format(score))
```

```python
# Classification Report
print(classification_report(ytrain, y_train_pred))
```
